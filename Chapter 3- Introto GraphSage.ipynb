```python
# imports
import torch
import torch.nn.functional as F
from torch_geometric.nn import SAGEConv
from torch_geometric.datasets import Planetoid
from torch_geometric.loader import DataLoader
```

```python
# model definition
class GraphSAGE(torch.nn.Module):
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.conv = SAGEConv(in_channels, out_channels)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv(x, edge_index)
        return F.log_softmax(x, dim=1)
```

```python
# dataset
dataset = Planetoid(root='/tmp/Cora', name='Cora')
```

* **Question 1 (1 pt):** Visualize the training data using a classical dimensionality reduction technique.

Use `device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')` if you'd like to use CUDA.

```python
# Setting up the model
model = GraphSAGE(dataset.num_node_features, dataset.num_classes)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)
```

```python
# Implement PCA to reduce dimensionality
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
data = dataset[0]
model.eval()
with torch.no_grad():
    out = model(data)
```

```python
pca = PCA(n_components=2)
reduced = pca.fit_transform(out)
```

```python
plt.scatter(reduced[:, 0], reduced[:, 1], c=data.y, cmap='viridis', s=30)
plt.colorbar(label='Class Labels')
plt.title("PCA Dimensional Reduction Visual")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.show()
```

* **Question 2 (1 pt):** What's "weight decay"?

"Weight decay" is a way to prevent overfitting by adding a penalty to the loss function. The penalty is imposed on larger weights in the data and allows the model to generalize better to the new model. It is also a way to perform L2 regularization. 

```python
# Defining a training step
def train():
    model.train()
    optimizer.zero_grad()
    out = model(dataset.data)
    loss = F.nll_loss(out[dataset.data.train_mask], dataset.data.y[dataset.data.train_mask])
    loss.backward()
    optimizer.step()

# Defining a test step
def test():
    model.eval()
    with torch.no_grad():
        out = model(dataset.data)
        pred = out.argmax(dim=1)
        flag_is_correct = pred[dataset.data.test_mask] == dataset.data.y[dataset.data.test_mask]
        acc = int(flag_is_correct.sum()) / int(dataset.data.test_mask.sum())
        return acc
```

```python
# Run one iteration
train()
acc = test()
print(f'Accuracy: {acc:.4f}')
```

```python
# Run for a few epochs:
accuracy = []
for epoch in range(100):
    train()
    accuracy.append(test())
    print(f'Accuracy on epoch {epoch+1}: {accuracy[-1]}')
```

```python
import pylab as pl
f, ax = pl.subplots(1, 1, figsize=(4, 2.5))
ax.plot(accuracy, '-')
ax.set_xlabel("Epoch", fontsize=13)
ax.set_ylabel("Accuracy", fontsize=13)
pl.show()
```

* **Question 3 (1 pt):** Is this model predicting all classes equally well? Demonstrate differences by using any well known way of quantifying accuracy across multiple classes.

```python
from sklearn.metrics import confusion_matrix

def get_report():
    model.eval()
    with torch.no_grad():
        out = model(dataset.data)
        pred = out.argmax(dim=1)
        flag_is_correct = pred[dataset.data.test_mask] == dataset.data.y[dataset.data.test_mask]
        acc = int(flag_is_correct.sum()) / int(dataset.data.test_mask.sum())
        return acc, pred[dataset.data.test_mask], dataset.data.y[dataset.data.test_mask]
```

* **Question 5 (2 pt):** Change the aggregation function to `max`. What's the default?

```python
class GraphSAGE5(torch.nn.Module):
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.conv = SAGEConv(in_channels, out_channels, aggr='max')

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv(x, edge_index)
        return F.log_softmax(x, dim=1)
    
model = GraphSAGE5(dataset.num_node_features, dataset.num_classes)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)
```

* **Question 6 (2 pt):** Add one more layer to the model and investigate the performance change.

```python
class GraphSAGE6(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels):
        super().__init__()
        self.conv1 = SAGEConv(in_channels, hidden_channels)
        self.conv2 = SAGEConv(hidden_channels, out_channels)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = self.conv2(x, edge_index)
        return F.log_softmax(x, dim=1)
```

```python
model = GraphSAGE6(dataset.num_node_features, 120, dataset.num_classes)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)
```

```python
# Run for a few epochs:
accuracy = []
for epoch in range(100):
    train()
    accuracy.append(test())
    print(f'Accuracy on epoch {epoch+1}: {accuracy[-1]}')
```

```python
import pylab as pl
f, ax = pl.subplots(1, 1, figsize=(4, 2.5))
ax.plot(accuracy, '-')
ax.set_xlabel("Epoch", fontsize=13)
ax.set_ylabel("Accuracy", fontsize=13)
pl.show()
```

I added another SAGEConv layer into the GraphSage model, along with another parameter to add a number of hidden channels. By adding a parameter of 1230 hidden channels and keeping the aggregation function as mean for both layers, I was able to observe a slight spike in performance, reaching an accuracy level of 80 percent.

* **Question 7 (2 pt):** Add drop-out to this NN and show the impact to performance. How does it change the ROC curve.


```python
class GraphSAGE7(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels):
        super().__init__()
        self.conv1 = SAGEConv(in_channels, hidden_channels)
        self.drop = torch.nn.Dropout(p=0.5)
        self.conv2 = SAGEConv(hidden_channels, out_channels)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = self.drop(x)
        x = self.conv2(x, edge_index)
        return F.log_softmax(x, dim=1)
```

```python
model = GraphSAGE6(dataset.num_node_features, 120, dataset.num_classes)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)
```

```python
# Run for a few epochs:
accuracy = []
for epoch in range(100):
    train()
    accuracy.append(test())
    print(f'Accuracy on epoch {epoch+1}: {accuracy[-1]}')
```

```python
import pylab as pl
f, ax = pl.subplots(1, 1, figsize=(4, 2.5))
ax.plot(accuracy, '-')
ax.set_xlabel("Epoch", fontsize=13)
ax.set_ylabel("Accuracy", fontsize=13)
pl.show()
```
